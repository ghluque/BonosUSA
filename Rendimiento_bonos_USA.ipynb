**Para importe datos desde direccion de enlace o url (archivo en formato .csv llamado dataset) de la pagina oficial de datos del estado**


```
df=pd.read_csv('https://infra.datos.gob.ar/catalog/sspm/dataset/132/distribution/132.1/download/rendimiento-bonos-estados-unidos-tasas-porcentaje-frecuencia-diaria.csv')
df.head()

```

**Definimos las columnas de datos que van a ir en cada eje, nombre, color de cada trazado (segun el plazo) y grosor. Cerramos con el nombre del grafico y mostramos**


```
fig = go.Figure()


fig.add_trace(go.Scatter(
                x=df.indice_tiempo,                
                y=df.titulos_us_1_anios,                
                name="Titulos USA 1 a√±o",
                line_color='darkviolet',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=df.indice_tiempo,              
                y=df.titulos_us_2_anios,                
                name="Titulos USA 2 a√±os",
                line_color='salmon',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=df.indice_tiempo,        
                y=df.titulos_us_3_anios,
                name="Titulos USA 3 a√±os",
                line_color='aquamarine',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=df.indice_tiempo,
                y=df.titulos_us_5_anios,
                name="Titulos USA 5 a√±os",
                line_color='orangered',
                opacity=0.8))
fig.update_layout(title_text="Rendimiento de inversi√≥n en bonos de la Reserva Federal")
fig.show()
```
